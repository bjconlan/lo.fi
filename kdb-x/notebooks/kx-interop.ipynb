```python
N = 100
qtable = kx.Table(data={
    'x': kx.random.random(N, ['a', 'b', 'c']),
    'x1': kx.random.random(N, 1.0),
    'x2': 5 * kx.random.random(N, 1.0)})
display(qtable[0:5])
display(kx.q.sql("SELECT * FROM $1 WHERE x='a'", qtable[0:5]))
qtable[0:5].select(where=kx.Column('x')=='a')
```

```python
qvec = kx.q.til(10)
display(qvec)
kx.q.mavg(3, qvec)
```
