```python
import pykx as kx

# Dividend adjusted daily quote
daily_quote = kx.schema.builder({
  'date': kx.DateAtom,
  'sym': kx.SymbolAtom,
  'open': kx.IntAtom,
  'high': kx.IntAtom,
  'low': kx.IntAtom,
  'close': kx.IntAtom,
  'volume': kx.IntAtom,
})

db = kx.DB(path='/tmp/db', overwrite=True)
```

```python

from pathlib import Path
import pandas as pd
import numpy as np
from datetime import date

for file_path in Path('/workspaces/lo.fi/data/asx_stock_20251226').glob('*.csv'):
  path = file_path.resolve()
  df = pd.read_csv(path, usecols=['date', 'open', 'high', 'low', 'close', 'volume'])
  df['date'] = pd.to_datetime(df['date']).dt.date
  df['open'] = (df['open'] * 1000).astype(np.int32)
  df['high'] = (df['high'] * 1000).astype(np.int32)
  df['low'] = (df['low'] * 1000).astype(np.int32)
  df['close'] = (df['close'] * 1000).astype(np.int32)
  df['volume'] = df['volume'].astype(np.int32)
  df['sym'] = path.stem
  db.create(daily_quote.insert(kx.toq.from_pandas_dataframe(df)), 'asx_daily', 'date', by_field = 'sym')
```
